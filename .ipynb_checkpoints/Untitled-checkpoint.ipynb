```python
import numpy as np
import pandas as pd
# content based Movie recomendation system
```

```python
movies=pd.read_csv('tmdb_5000_movies.csv')
credits=pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies.head(1)
```

```python
credits.head(1)
```

```python
movies=movies.merge(credits,on='title')
```

```python
movies.head(1)
```

```python
movies.shape
```

```python
#geners
# id
# keyword
# title
# overview
# cast
# crew

movies=movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.head()
```

```python
movies.info()
```

```python
# Data Preprocessing
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.isnull().sum()
# all mising values we drop
```

```python
movies.duplicated().sum()
# no duplicate movies in our dataset
```

```python
movies.iloc[0].genres
```

```python
# in that we only use name so we remove id from geners 
import ast   
# it can convert dictionery to list
# now we convert the genres through object
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L;
        
```

```python
movies['genres'].apply(convert)
# here we ca get all the geners in that movie
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
# now we apply same thing on keywords
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
# now it cast we only consider top 3 actors . so we make a list on top 3 actors in cast
# so in cast column we only have to retrive the original name of the actors
def convert3(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj):
        if (counter != 3):
             L.append(i['name'])
             counter+=1
        else:
            break
    return L
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
# now for crew we only retrive the data that whose job is director
def fetch_director(obj):
        L=[]
        for i in ast.literal_eval(obj):
            if i['job']=='Director':
                L.append(i['name'])
                break
        return L
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
movies.head()
# now we fatch the directors from crew
```

```python
# now we convert the overview tupple from string to list , so we can easily concanitate with geners,keyword,cast and crew
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
# now we remove the blank spaces beetween the 2 words like sam worthington to samworthington
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x]) 
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
# now we concnitate genres,keywords,cast,crew
movies['tags']=movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
new_df=movies[['movie_id','title','tags']]
```

```python
new_df.head()
```

```python
# now we convert the tags from list to string
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.head()
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

```python
# now we use the consept of vectorization 
# in vectorization we split the tags into words and then into vectors
# now when we select one movie then we recomend closest 5 vectors 
```

```python
# now we concanitate all the tags and we fatch 5000 common words from them 
# now we check the frequency of this words with indiviual tags 
# whose frequency is high we reccomend them
# here we not consider sentense formation words
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors.shape
```

```python
cv.get_feature_names()
# we can see this 5000 words
```

```python
# we can see that some are common mining words are there like actor and actors , activiy and activities 
# so we have to remove this words
# so we can use steming to avoid this problem
```

```python
!pip install nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return  " ".join(y)
```

```python
ps.stem('loving','loved')
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
new_df['tags']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
cv.get_feature_names()
```

```python
# now we find cosine distanse of every movie and which movies have less distanse we can recommend them
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vectors)
```

```python
similarity=cosine_similarity(vectors)
```

```python
similarity
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title']==movie].index[0]
    distances=similarity[movie_index]
    movie_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    
    for i in movie_list:
        print(new_df.iloc[i[0]].title)
        
     
```

```python
recommend('Avatar')
```

```python
recommend('Batman Begins')
```
